```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("netflix").getOrCreate()
```

```python
df = spark.read.csv(cur_path + 'netflix_titles.txt',inferSchema=True,header=True,sep='\t')
```

```python
df.printSchema()
```

```python
import itertools
from pyspark import SparkContext
```

```python
df = df.na.drop(subset=['director'])
```

```python
df = df.na.drop(subset=['cast'])
```

```python
df2 = df.select(['director','cast'])
```

```python
data = df2.toPandas()
```

```python
data.head()
```

```python
data.shape
```

```python
data = data.assign(director=data.director.str.split(", ")).explode('director')
```

```python
data = data.assign(cast=data.cast.str.split(", ")).explode('cast')
```
